```python
import numpy as np
import random
import matplotlib.pyplot as plt

N = 9999#采样点数
delta = 1

def g(x):
    return np.exp(-x)
def h(x):
    return np.power(x,3/2)

#Metropolis抽样
x_0 = 1e-5
sample = []
while len(sample)<=N:
    x_1 = x_0+2*delta*random.random()-delta
    while(x_1<=0):
        x_1 = x_0+2*delta*random.random()-delta
    r = g(x_1)/g(x_0)
    if r>random.random():
        x_0 = x_1
        continue
    sample.append(x_0)

#积分
sum_ = 0
integral = 0
for x in sample:
    sum_ += h(x)
integral = sum_/len(sample)
print(integral)

#抽样分布图
plt.hist(sample, 100, density=True, color='red', label='samples distribution')
plt.xlabel('sample')
plt.legend()
plt.show()
```
