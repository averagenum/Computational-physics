```python
import numpy as np
import random
import matplotlib.pyplot as plt

#parameter
N_L = 10000
N_R = 0
Total = N_L+N_R
S = 30000 # Total steps
step = 0

#The transfer probability, assuming the probability is proportional to the number of molecules.
def p(N1,N2):
    return N1/(N1+N2)
#another p for test
def p_same(a,b):
    return 0.5

#Record the molecular number changes after each step.
L = []
R = []
L.append(N_L)
R.append(N_R) 
```

```python
while(step <=S):
    if(random.randint(1,Total)<=N_L):
        #the moleculer is in L
        if(random.random()<=p(N_L,N_R)):
            #transfer to R
            N_L -= 1
            N_R += 1
    else:
        #the moleculer is in R
        if(random.random()<=p(N_R,N_L)):
            N_R += -1
            N_L += 1
    L.append(N_L)
    R.append(N_R)
    step += 1
        
plt.plot(L, label="L")
plt.plot(R, label="R")
plt.xlabel('times')
plt.ylabel('moleculer number')
plt.legend()
plt.show()
```
